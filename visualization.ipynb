```python
import pandas as pd 
# Load the data from csv
df = pd.read_csv('car_buying.csv')
df.head()
```

```python
df['BoughtCar'] = df['BoughtCar'].map({'Yes': 1, 'No': 0})
df.head()
```

```python
X = df[['Age', 'Income', 'CreditScore']] # Featutres
Y =df['BoughtCar'] # Target Variable
```

```python
from sklearn.model_selection import train_test_split

# split the data into training and testing sets (80% train, 20% test)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# check the shape of the data
print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")
```

```python
from sklearn.tree import DecisionTreeClassifier


# Create and train the model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, Y_train)
```

```python
# check the accuracy of the  model on the test data
accuracy = model.score(X_test, Y_test)
print(f'Accuracy on the test data: {accuracy:.2f}%')
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Plot the dicision tree
plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=X.columns, class_names=['No', 'Yes'], filled=True)
plt.title('Decision Tree for Car Buying Decision')
plt.show()

# Save the model using joblib
```
